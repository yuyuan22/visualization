```python
from pandas import read_csv
from matplotlib import pyplot
from pandas.plotting import lag_plot
series = read_csv('daily-min-temperatures.csv', header=0, index_col=0, parse_dates=True)
lag_plot(series)
pyplot.show()
```

```python
series
```

```python
series['Temp_Previous'] = series['Temp'].shift(1)
```

```python
series.plot(kind='scatter',x='Temp', y='Temp_Previous')
```

```python
series.plot(kind='hexbin', x='Temp', y='Temp_Previous', gridsize = 13)
```

```python
import pandas as pd
from random import random
  
# initialize list of lists
data = []
for i in range(21):
    tmp = [2010 + i, 2000 + i * 100  + 200 *random(), 'Team A']
    tmp2 = [2010 + i, 1200 + i * 100 +  200 * random(), 'Team B']
    data.append(tmp)
    data.append(tmp2)
    
# Create the pandas DataFrame
df = pd.DataFrame(data, columns=['Year', 'Revenue', 'Team'])
  
# print dataframe.
df
```

```python
import plotly.express as px

fig = px.bar(df, x="Year", y="Revenue", color="Team")
fig.update_layout(barmode='group')
fig.show()
```

```python
fig = px.scatter(df, x="Year", y="Revenue", color="Team")

fig.show()
```

```python
import pandas as pd
from random import randint
  
# initialize list of lists
data = []
for i in range(11):
    tmp = [str(2010 + i), 200 + i * 10  + randint(0,40)-40]
    if i < 1:
        tmp.append(tmp[-1])
    else:
        tmp.append(tmp[-1] - data[-1][1])
    data.append(tmp)
    
# Create the pandas DataFrame
df = pd.DataFrame(data, columns=['Year', 'Value', 'Delta'])
  
# print dataframe.
df
```

```python
df.plot.line("Year", "Value")
```
